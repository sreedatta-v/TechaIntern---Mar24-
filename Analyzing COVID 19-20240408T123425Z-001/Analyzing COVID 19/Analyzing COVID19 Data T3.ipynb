```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('owid-covid-data.csv')
```

```python
data.head()
```

```python
data.dtypes
```

```python
print(data.shape)
```

```python
data.isnull().sum()
```

```python
data = data[['date', 'total_cases', 'total_deaths']]
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
sample_data = data.sample(n=100)
plt.plot(sample_data['date'], sample_data['total_cases'])
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.title('Daily New COVID-19 Cases')
plt.show()
```

```python
max_values = data[['total_cases', 'total_deaths']].max()
max_values.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Maximum Cases/Deaths/Recoveries')
plt.title('Comparison of Maximum Values Across Countries')
plt.show()
```
